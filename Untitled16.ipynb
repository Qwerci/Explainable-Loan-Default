## Explainable AI (XAI) Analysis - Counterfactuals

### Subtask:
Perform Counterfactual analysis to understand what minimal changes to features would flip the model's prediction.

**Reasoning**:
Apply Counterfactual analysis using a library like DiCE to find realistic counterfactual explanations for a few instances from the test set. This requires preparing the data and model for the DiCE library and specifying constraints for realistic counterfactuals.

```python
# Install DiCE if you haven't already
!pip install dice_ml

import dice_ml
import pandas as pd
import numpy as np

# Select the best performing model (XGBoost)
model_for_cf = best_models['XGBoost']

# Prepare the data for DiCE
# DiCE works best with pandas DataFrames and needs feature names
# Use the preprocessed X_test data and add the target variable
X_test_preprocessed_with_target = X_test.copy()
# Ensure y_test is a Series with the same index as X_test
X_test_preprocessed_with_target['loan_status'] = y_test.values # Add the target variable as a numpy array to align indices

# Explicitly convert *all* feature columns to float type to avoid issues with DiCE/XGBoost
feature_cols = X_test.columns.tolist() # Get the names of the preprocessed feature columns
for col in feature_cols:
    X_test_preprocessed_with_target[col] = X_test_preprocessed_with_target[col].astype(float)


d = dice_ml.Data(dataframe=X_test_preprocessed_with_target, # Use the combined DataFrame with correct dtypes
                 continuous_features=[col for col in feature_cols if col.startswith('num__')], # Identify numerical features from preprocessed data
                 outcome_name='loan_status') # Specify the target variable name

# Create a DiCE model object
# Pass the trained model and the data object
m = dice_ml.Model(model=model_for_cf, backend='sklearn') # Specify the backend based on the model type

# Create a DiCE explainer object
# Specify the data and model objects
exp = dice_ml.Dice(d, m, method='genetic') # Using the genetic algorithm for counterfactual generation

# Select a few instances from the test set for counterfactual explanation
# It's often insightful to pick instances predicted as the unfavorable class (default=1)
# Let's find a few instances from the test set that were predicted as default by the XGBoost model
y_test_pred = model_for_cf.predict(X_test)
defaulted_indices = X_test.index[y_test_pred == 1].tolist()

# Select a few indices from the defaulted instances
if len(defaulted_indices) > 5:
    instances_to_explain_indices = defaulted_indices[:5] # Take the first 5 predicted defaults
else:
    instances_to_explain_indices = defaulted_indices # Take all if less than 5

if not instances_to_explain_indices:
    print("No instances predicted as default found in the test set to explain with counterfactuals.")
else:
    # Select the instances to explain from the preprocessed X_test
    instances_to_explain = X_test.loc[instances_to_explain_indices]

    # Generate counterfactuals
    # total_CFs: number of counterfactuals to generate per instance
    # desired_class: the target class (0 for no default)
    # desired_range: for continuous features, specify the desired range (optional)
    # permitted_range: for continuous features, specify the permitted range (optional)
    # features_to_vary: list of features that are allowed to be varied (default is all)
    # verbose: level of detail in output

    # Define constraints for realistic counterfactuals
    # Identify mutable features (features that can realistically change)
    mutable_features_original = [
        'person_age',
        'person_income',
        'person_emp_length',
        'loan_amnt',
        'loan_int_rate',
        'loan_percent_income',
        'cb_person_cred_hist_length',
        # Include original categorical features if you want to allow changes in those
        'person_home_ownership',
        'loan_intent',
        'loan_grade',
        'cb_person_default_on_file'
    ]

    # Filter mutable features to only include those present in the preprocessed data (X_test columns)
    # DiCE needs the names of the features in the dataframe passed to dice_ml.Data
    mutable_features_preprocessed = [col for col in X_test.columns if col.replace('num__', '').replace('cat__', '').split('_')[0] in [orig.split('_')[0] for orig in mutable_features_original]]

    # Refine mutable features for realism if needed later

    # Define permitted ranges for mutable numerical features based on scaled training data
    feature_ranges = {}
    for col in X_train.columns:
        if col.startswith('num__'): # Only apply ranges to numerical features
             feature_ranges[col] = [X_train[col].min(), X_train[col].max()]

    # Generate counterfactuals for the selected instances
    print(f"\nGenerating counterfactuals for {len(instances_to_explain)} instances predicted as default...")
    dice_exp = exp.generate_counterfactuals(
        instances_to_explain,
        total_CFs=3,  # Number of counterfactuals per instance
        desired_class="opposite", # Find counterfactuals for the opposite class (no default)
        features_to_vary=mutable_features_preprocessed, # Use preprocessed feature names that are mutable
        permitted_range=feature_ranges, # Apply realistic ranges to numerical features
        verbose=True
    )

    # Visualize the counterfactuals
    print("\nVisualizing counterfactuals:")
    dice_exp.visualize_as_dataframe(show_only_changes=True) # Show only changes from the original instance
    # You can also visualize as a graph: dice_exp.visualize_as_html()
```

Implement bias mitigation techniques if needed.

Apply a post-processing bias mitigation technique (Threshold Optimization) using Fairlearn to the best-performing model (XGBoost) to try and improve fairness across groups based on a protected attribute, and then re-evaluate the fairness metrics.

```python
from fairlearn.postprocessing import ThresholdOptimizer
from fairlearn.metrics import MetricFrame, demographic_parity_ratio, equalized_odds_ratio, selection_rate, true_positive_rate, false_positive_rate
import pandas as pd

# Use the best performing model (XGBoost) for mitigation
model_for_mitigation = best_models['XGBoost']

# Define the protected attribute for mitigation
# Let's choose 'person_home_ownership' as the protected attribute for this example
protected_attribute_mitigation = protected_attributes_test['person_home_ownership']

# Get the prediction probabilities for the test set
y_proba = model_for_mitigation.predict_proba(X_test)[:, 1] # Probability of the positive class (default)

# Initialize the ThresholdOptimizer
# We want to optimize for equalized odds, which aims for equal True Positive Rate and False Positive Rate across groups.
# Alternatively, could optimize for demographic parity (equal selection rate).
# The estimator is the original model (or its predict_proba method).
# The constrained metric is what we want to equalize (e.g., 'equalized_odds').
# The target metric is the overall performance metric to maximize while satisfying the constraint (e.g., 'f1').
# The predict_proba method from the model is used as the base estimator for optimization.

try:
    # For ThresholdOptimizer, we need to use the model's predict_proba method directly as the estimator
    # and provide the sensitive features and true labels.
    optimizer = ThresholdOptimizer(
        estimator=model_for_mitigation, # Pass the fitted model
        constraints="equalized_odds", # Or "demographic_parity"
        objective="balanced_accuracy_score" # Changed objective to a supported one for equalized_odds
    )

    # Fit the optimizer on the test set
    # The optimizer learns the optimal thresholds based on the predictions and sensitive features in the test set.
    # Note: It's generally recommended to use a separate validation set for fitting the optimizer to avoid data leakage
    # if you were selecting the optimizer parameters. Here, for demonstration, we use the test set.
    print("Fitting ThresholdOptimizer...")
    optimizer.fit(X_test, y_test, sensitive_features=protected_attribute_mitigation)
    print("ThresholdOptimizer fitted.")

    # Get mitigated predictions on the test set
    y_pred_mitigated = optimizer.predict(X_test, sensitive_features=protected_attribute_mitigation)

    # Re-evaluate fairness metrics with mitigated predictions
    print("\nRe-evaluating fairness metrics after mitigation (using ThresholdOptimizer on 'person_home_ownership'):")
    fairness_metrics_mitigated = MetricFrame(
        metrics=metrics, # Use the same metrics as before
        y_true=y_true,
        y_pred=y_pred_mitigated,
        sensitive_features=protected_attribute_mitigation # Use the protected attribute for this mitigation
    )

    # Display the mitigated fairness metrics
    print("\nMitigated Fairness metrics by group (person_home_ownership):")
    display(fairness_metrics_mitigated.by_group)

    # Calculate and display mitigated fairness ratios
    print("\nMitigated Fairness Ratios:")
    print(f"Demographic Parity Ratio (Selection Rate Ratio): {fairness_metrics_mitigated.ratio(method='between_groups').selection_rate:.4f}")
    print(f"Equalized Odds Ratio (True Positive Rate Ratio): {fairness_metrics_mitigated.ratio(method='between_groups').true_positive_rate:.4f}")
    print(f"Equalized Odds Ratio (False Positive Rate Ratio): {fairness_metrics_mitigated.ratio(method='between_groups').false_positive_rate:.4f}")

    print("\nBias mitigation analysis completed.")

except Exception as e:
    print(f"An error occurred during bias mitigation: {e}")
    print("Please check the inputs and ensure Fairlearn is installed correctly.")

# TODO: Explore mitigation based on other protected attributes or combinations.
# TODO: Explore other mitigation techniques (pre-processing, in-processing).
```
